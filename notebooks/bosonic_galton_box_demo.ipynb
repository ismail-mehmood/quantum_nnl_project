## Classical Boson Sampler (Unitary Approach)

Inspired by quantum photonic systems, this classical simulation models a particle passing through a series of beam splitters (unitary layers), with amplitudes interfering constructively or destructively at output modes.

Each layer corresponds to a beam splitter array, and the output distribution can be tuned via the reflection coefficient `R`.

Though classical, this method provides a useful benchmark for photonic-inspired Galton board behaviors.

We can input a number of photons and a board depth to the function sample_basic_board() to simulate and return a plot of the binning outcomes.
